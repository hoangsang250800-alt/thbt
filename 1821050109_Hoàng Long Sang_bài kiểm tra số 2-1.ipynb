### Bài 1:

```python
n = int(input("Nhập n: "))
tong = 0
dem = 0

for i in range(n):
    x = float(input("Nhập số: "))
    if 0 < x < 1000:
        tong += x
        dem += 1

if dem > 0: 
    print("Trung bình cộng:", tong / dem)
```

### Bài 2

```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
temp=n
while temp > 0:
    chu_so= temp % 10    
    tong= tong + chu_so  
    temp= temp // 10    
if tong % 3 == 0:
    print("Số nguyên n=",n,"Chia hết cho 3")
else:
    print("Số nguyên n=",n,"Không chia hết cho 3")
```

### Bài 3

```python
n = int(input("Nhập so nguyen duong n: "))

tich = 1
temp = n 
while temp > 0:
    chu_so = temp % 10         
    tich = tich * chu_so        
    temp = temp // 10           
if tich % 2 == 0 and tich > 20:
    print("Thỏa mãn điều kiện", n)
else:
    print("Số n=",n,"không thỏa mãn !")
```

## Bài 4

```python
a = int(input("Nhập số  a: "))
b = int(input("Nhap số b: "))
tong = a + b
so_lon_nhat = 0
temp = tong

while temp > 0:
    chu_so = temp % 10
    if chu_so > so_lon_nhat:
          so_lon_nhat = chu_so
    temp = temp // 10
print("tong a+b:",tong)
print("Chữ số  lớn nhất trong dãy kq=",tong,"Suy ra số lớn nhất trong dãy kq là", so_lon_nhat)
```
