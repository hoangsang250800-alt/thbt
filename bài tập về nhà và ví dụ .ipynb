## bài tập 1: Nhập vào 3 điểm toán lý hóa. Tính điểm trung bình , điểm tb <5 thì yếu, điểm tb<7 trung bình <9 thì khá, còn lại là giỏi

```python

toan = int(input("Nhap diem Toan: "))
ly = int(input("Nhap diem Ly: "))
hoa = int(input("Nhap diem Hoa: "))

dtb = (toan + ly + hoa) / 3
print("Diem trung binh la:", dtb)

if dtb < 5:
    print("Xếp loại: Yếu!")
elif dtb < 7:
    print("Xếp loại: Trung bình!")
elif dtb < 9:
    print("Xếp loại: Khá!")
else:
    print("Xếp loại: Giỏi!")
```

## bài tập 2: nhập vào một dãy số tính tổng các số có mặt trong dãy số đó

 ## nhập N viết hàm tính giai thừa số đó

```python
def giaithua(m):
    gt = 1
    for i in range(1, m + 1):
        gt = gt * i 
    return gt  
n = int(input("Nhập vào 1 số nguyên dương: "))
print("%d! = %d" % (n, giaithua(n)))
```
