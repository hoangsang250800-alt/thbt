```python

toan = int(input("Nhap diem Toan: "))
ly = int(input("Nhap diem Ly: "))
hoa = int(input("Nhap diem Hoa: "))

dtb = (toan + ly + hoa) / 3
print("Diem trung binh la:", dtb)

if dtb < 5:
    print("Xếp loại: Yếu!")
elif dtb < 7:
    print("Xếp loại: Trung bình!")
elif dtb < 9:
    print("Xếp loại: Khá!")
else:
    print("Xếp loại: Giỏi!")
```

 ## nhập N viết hàm tính giai thừa số đó

```python
def giaithua(m):
    gt = 1
    for i in range(1, m + 1):
        gt = gt * i 
    return gt  
n = int(input("Nhập vào 1 số nguyên dương: "))
print("%d! = %d" % (n, giaithua(n)))
```

## Viết chương trình nhập vào một dãy số thực x1, x2, …, xn (0 < n < 100), sau đó tìm trung bình cộng các phần tử âm trong dãy mà giá trị nằm trong khoảng (-1000, -10)

```python
n = int(input("Nhập số lượng phần tử n : "))
day_so = list(map(float, input("Nhập dãy số thực: ")))
tam = []
for x in day_so:
    if -1000 < x < -10:
        tam.append(x)       
a = len(tam)
if a > 0:
    tbc = sum(tam) / a
    print("Các số nhặt được:", tam)
    print(" Trung bình cộng là:", tbc)
else:
    print(" Không có số âm nào nằm trong khoảng (-1000, -10)!")
```

## Viết chương trình nhập vào một dãy số nguyên x1, x2, …, xn (0 < n < 200), tính tổng các phần tử chẵn trong dãy, và kiểm tra xem tổng này có chia hết cho 7 và nhỏ hơn 200 hay không


```python
n = int(input("Nhập số lượng phần tử n (0 < n < 200): "))
day_so = list(map(int, input("Nhập dãy số nguyên: ")))
tam = []
for x in day_so:
    if x % 2 == 0:  
        tam.append(x)  
tong = sum(tam)
print("Các số chẵn nhặt được là:", tam)
print("Tổng các số chẵn là:", tong)
if tong % 7 == 0 and tong < 200:
    print("Thỏa mãn")
else:
    print("Không thỏa mãn")
```
